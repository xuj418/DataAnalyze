# 4.2.4 時系列データ

## 一ヶ月分のデータを作る

```python
#freq='D'
dates = pd.date_range(start='2020/04/01', end="2020/04/30")
dates
```

```python
np.random.seed(666)
monthDate = pd.DataFrame(np.random.randint(1,31,30), columns=["乱数"], index=dates)
```

```python
monthDate
```

## 一年分のデータを作る

```python
dates = pd.date_range(start='2020/01/01', periods=366)
dates
```

```python
np.random.seed(666)
yearDate = pd.DataFrame(np.random.randint(1,31,366), columns=["乱数"], index=dates)
```

```python
yearDate
```

## 月平均のデータにする

```python
#method1:groupby
#pd.Grouper(freq='M'):月によりグルーピング
yearDate.groupby(by=pd.Grouper(freq='M')).mean()
```

```python
pd.Grouper(freq='M')
```

```python
#method2:resample
#resample:再度sampling,rule=M,D
yearDate.loc[:, "乱数"].resample(rule="M").mean()
```

## 複雑なインデックス

```python
#freq="W-SAT","D","M"
pd.date_range(start="2020/1/1", end="2020/12/31", freq="W-SAT")
```

```python
yearDate.groupby(by=pd.Grouper(freq="W-SAT")).sum()
```

# 4.2.5 欠損値処理

```python
df05 = pd.read_csv("data/201705health.csv")
```

```python
df05
```

```python
#dropna:欠損値を削除
df05_drop = df05.dropna()
df05_drop
```

```python
#fillna:欠損値に０を代入
df05_fill = df05.fillna(value="0.0")
df05_fill
```

```python
#method="ffill":一つ手前の値で補完
df05_fill_method = df05.fillna(method="ffill")
df05_fill_method
```

```python
#平均値で補完
df05_fill_value_mean = df05.fillna(value=np.mean(df05))
df05_fill_value_mean
```

```python
#中央値で補完
df05_fill_value_median = df05.fillna(value=df05.median())
df05_fill_value_median
```

```python
#np.median:(ERROR:not supported between instances of 'float' and 'str')
#df05_fill_value_median = df05.fillna(value=np.median(df05))
```

```python
#df05.mode():最頻値を返す
df05_fill_value_fre = df05.fillna(value=df05.mode().iloc[0,:])
df05_fill_value_fre
```
