## タイトル等

```python
matplotlib.style.use("ggplot")
fig, axes = plt.subplots(ncols=2)


for i in range(len(axes)):
    axes[i].set_title("Title{}".format(i))   #set_title：サブプロットにタイトルを設定
    axes[i].set_xlabel("XLabel{}".format(i))  #set_xlabel:軸ラベルを設定
    axes[i].set_ylabel("YLabel{}".format(i))
    axes[i].plot(x, y, label="XX Test{}".format(i))   #label：凡例を設定
    axes[i].legend(loc="best")   #legend：凡例を表示

# suptitle：描画オブジェクトにタイトルを設定
fig.suptitle("figure title")
fig.savefig("data/pngg.png")    #savefig:ファイル出力
plt.show()
```

# 4.3.3 グラフの種類と出力方法

## 折れ線グラフ

```python
fig, ax = plt.subplots()

ax.plot([1,2,3], [1,2,3], label="a1")
ax.plot(x, y, label="a2")
ax.legend(loc="best")
plt.show()
```

```python
x = np.linspace(0, 15, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label="sinx")
ax.plot(x, y2, label="cosx")
ax.legend(loc="best")

plt.show()
```

## 棒グラフ

```python
x = np.array([1,2,3])
y = np.array([3,1,2])
labels = np.array(["apple", "banana", "orange"])
#y = [3,1,2]  #ERROR
fig, axes = plt.subplots(2)
axes[0].bar(x, y, color="b", tick_label=labels)  #bar:縦向き
axes[1].barh(x, y, color="c", tick_label=labels)  #barh:横向き

plt.show()
```

```python
x = np.array([1, 2, 3])
y1 = np.array([3, 1, 2])
y2 = np.array([4, 1, 4])

labels = np.array(["apple", "banana", "orange"])
fig, ax = plt.subplots()

width = 0.3   #width：棒グラフの幅を設定
ax.bar(x, y1, color="b", width=width, tick_label=labels, label="y1")

x2 = np.array([i + width for i in x])   #幅分ずらして描画
ax.bar(x2, y2, color="g", width=width, label="y2")

ax.legend(loc="best")
plt.show()
```

```python
x = np.array([1, 2, 3])
y1 = np.array([3, 1, 2])
y2 = np.array([4, 1, 4])
labels = np.array(["apple", "banana", "orange"])

y_total = y1 + y2

fix, ax = plt.subplots()

width=0.5

#二つの積み上げ棒グラフ
#まず、全体の棒グラフを描画
ax.bar(x, y_total, width=width, tick_label=labels, label="y2")
#次、y１の棒グラフを描画
ax.bar(x, y1, width=width, label="y1")

ax.legend()
plt.show()
```

```python
x = np.array([1, 2, 3])
y1 = np.array([3, 1, 2])
y2 = np.array([4, 1, 4])
y3 = np.array([5, 5, 5])
labels = np.array(["apple", "banana", "orange"])

y_total2 = y1 + y2 + y3
y_total1 = y1 + y2

fix, ax = plt.subplots()

width=0.5

#三つの積み上げ棒グラフ
ax.bar(x, y_total2, width=width, tick_label=labels, label="y3")
ax.bar(x, y_total1, width=width, label="y2")
ax.bar(x, y1, width=width, label="y1")

ax.legend()
plt.show()
```

## 散布図

```python
fig, ax = plt.subplots()

x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y)    #scatter:散布図を生成
plt.show()
```

```python
fig, ax = plt.subplots()

np.random.seed(666)
x = np.random.rand(50)
y = np.random.rand(50)

#o、+、＾：上向き三角形、V：下向き三角形、<：左向き三角形、>：左向き三角形
#s：正方形、*：星、x、_
ax.scatter(x[:10], y[:10], marker="o")
ax.scatter(x[10:20], y[10:20], marker="+")
ax.scatter(x[20:30], y[20:30], marker="^")
ax.scatter(x[30:40], y[30:40], marker=">")
ax.scatter(x[40:50], y[40:50], marker="_")

plt.show()
```

## ヒストグラム

```python
#ヒストグラム：連続データの分布特徴を描画
avg = 100
sigma = 15
np.random.seed(666)
x = np.random.normal(avg, sigma, 100)

fig, ax = plt.subplots()

#n:各棒の要素数,bins:各棒の境界値,patches:各棒の情報
n, bins, patches = ax.hist(x, color="g")   #hist:ヒストグラムを描画

plt.show()
```

```python
print(n)
print(bins)
print(patches[0])
```

```python
for i, num in enumerate(n):
    print("{:.2f} - {:.2f} : {}".format(bins[i], bins[i+1], num))
```

```python
avg = 100
sigma = 15
np.random.seed(666)
x = np.random.normal(avg, sigma, 100)

fig, axes = plt.subplots(3)
axes[0].hist(x, color="g")
axes[1].hist(x, color="g", bins=5)      #bins:棒の数を変更、ディフォトは１０
axes[2].hist(x, color="g", range=(80,100))      #range:範囲を設定

plt.show()
```

## 箱ひげ図

```python
avg = 100
np.random.seed(666)
x0 = np.random.normal(avg, 10, 1000)
x1 = np.random.normal(avg, 20, 1000)
x2 = np.random.normal(avg, 30, 1000)
label = np.array(["x0", "x1", "x2"])

fig, ax = plt.subplots()
#上から：上限、上四分位数、中位数、下四数位数、下限；
#丸：異常値
ax.boxplot((x0, x1, x2), labels=label)  #boxplot:箱ひげを描画
#ax.legend()

plt.show()
```

## 丸グラフ

```python
x = np.array([10, 3, 2])
label = np.array(["x0", "x1", "x2"])

fig, ax = plt.subplots()
ax.pie(x, labels=label)

plt.show()
```

```python
x = [10, 3, 2]
label = ["x0", "x1", "x2"]

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,10))  #figsize:サイズを設定
axes[0, 0].pie(x, labels=label)
axes[0, 0].set_title("basic")
axes[0, 1].pie(x, labels=label, startangle=90)  #startangle:start angle開始角度を設定
axes[0, 1].set_title("startangle")
axes[1, 0].pie(x, labels=label, autopct="%.2f%%")  #autopct:値のパーセントを設定
axes[1, 0].set_title("autopct")
axes[1, 1].pie(x, labels=label, counterclock=False)  #counterclock:時計回りに配置、False：時計周り
axes[1, 1].set_title("counterclock")
axes[2, 0].pie(x, labels=label, shadow=True)   #shadow:影をつける
axes[2, 0].set_title("shadow")
axes[2, 1].pie(x, labels=label, explode=[0.3,0.2,0])   #explode:一部の値を目立たせる
axes[2, 1].set_title("explode")

plt.show()
```

```python
np.random.seed(999)
num = 1000
x = np.random.randn(num)

fig, ax = plt.subplots()
counts, edges, patches = ax.hist(x, color="g", bins=20)

x_plot = (edges[:-1] + edges[1:]) / 2
y_plot = num *(np.diff(edges)) * (np.exp(-(x_plot ** 2) / 2) / np.sqrt(2 * np.pi))
plt.plot(x_plot, y_plot)

plt.show()
```

# 4.3.4 スタイル

## 色の設定

```python
x = np.arange(1, 4)
y1 = np.arange(2, 5)
y2 = np.array([2, 2, 2])
y3 = np.array([4, 3, 2])

fig, ax = plt.subplots()
#method01
ax.plot(x, y1, color="green", label="green")
#method02 16進数のRGB
ax.plot(x, y2, color="#0000FF", label="blue")
#method03　RGBをFloatで指定
ax.plot(x, y3, color=(0,0,0), label="black")

ax.legend()
plt.show()
```

```python
x = [1, 3]
y = [2, 2]

fig, ax = plt.subplots()
ax.bar([1], [2], color="green", edgecolor="red")  #edgecolor:境界の色
ax.bar([2], [2], color="white", edgecolor="blue")

plt.show()
```

## 線のスタイル

```python
x = np.arange(1, 4)
y1 = np.arange(2, 5)
y2 = np.array([2, 2, 2])
y3 = np.array([4, 3, 2])

fig, axes = plt.subplots(nrows=2, figsize=(5,5))
#linewidth
axes[0].plot(x, y1, color="green", linewidth=5, label="lw5")
axes[0].plot(x, y2, color="#0000FF", linewidth=10, label="lw10")
axes[0].plot(x, y3, color=(0,0,0), linewidth=15, label="lw15")
axes[0].legend()

#linestyle
axes[1].plot(x, y1, color="green", linewidth=3, linestyle=":", label="dotted")
axes[1].plot(x, y2, color="#0000FF", linewidth=3, linestyle="--", label="dashed")
axes[1].plot(x, y3, color=(0,0,0), linewidth=3, linestyle="-.", label="dashdot")
axes[1].legend()

plt.show()
```

## フォント

```python
fig, ax = plt.subplots()
ax.set_xlabel("xlabel", size=20, weight="bold", family="fantasy")
ax.set_ylabel("ylabel", size=30, weight="light", family="cursive")
ax.set_title("font", size=20, weight="normal", family="monospace")

plt.show()
```

```python
# フォントのスタイルを指定
fontdict = {
    "size" : 20,
    "weight" : "normal",
    "family" : "monospace"
}
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("xlabel", fontdict=fontdict)
ax.set_ylabel("ylabel", fontdict=fontdict)
ax.set_title("font", fontdict=fontdict, size=40)

plt.show()
```
